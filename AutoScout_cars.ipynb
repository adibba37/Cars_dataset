```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=(10,6)
```

```python
# Load the data
cars = pd.read_csv('https://bit.ly/3q5Fi81', parse_dates=['Year'])
```

```python
# Print the first five rows
cars.head()
```

```python
# Print the last five rows
cars.tail()
```

```python
# Get the # of rows and columns
cars.shape
```

```python
# Get the column names
cars.columns
```

```python
# Make type
cars.Make.unique()
```

```python
# Model type
print(cars.Model.unique())
print("# of Model:",cars.Model.nunique())
```

```python
# Door type
cars.Doors.unique()
```

```python
# Kind type
cars.Kind.unique()
```

```python
# Unique Year
cars.Year.unique()
```

```python
# Convert doors to int
cars['Doors'] = cars['Doors'].astype(int)

# Current year
current_year = datetime.now().year

# Convert 'Year' to integer and calculate age
cars['Age'] = current_year - cars['Year'].dt.year
```

```python
# Get the info about the data
cars.info()
```

```python
cars.isna().sum()
```

```python
# Basic statistics
cars.describe().round().transpose()
```

```python
# Most expensive car in the dataset
cars[ cars['Price']==cars['Price'].max()]
```

```python
# Most chepeast car in the dataset
cars[ cars['Price']==cars['Price'].min()]
```

```python
# Most expensive car type in the dataset
cars.Kind[ cars.Price == cars.Price.max()]
```

```python
# Cheapest and most Expensive car
cars[ (cars['Price'] == cars['Price'].min()) | (cars['Price'] == cars['Price'].max())] 
```

```python
# Pair plot
sns.pairplot(cars[['Make','Year', 'Kms', 'Price','Age']], hue='Make')
plt.show()
```

```python
# Histogram for the 'Price'
sns.displot(cars['Price'], bins=20, kde=True) 
plt.title('Distribution of Price')
plt.show()
```

```python
# Histogram for the 'Kms' 
sns.displot(cars['Kms'], bins=20, kde=True)
plt.title('Distribution of Kms')
plt.show()
```

```python
# Histogram for the 'Age' 
sns.kdeplot(cars['Age'])
plt.title('Distribution of Age')
plt.show()
```

```python
# Kdeplot for Year
sns.kdeplot(cars['Year'])
plt.title('KDE Plot for Car Year Distribution')
plt.show()
```

```python
# Box plot 'Make' and 'Price'
sns.boxplot(x='Make', y='Price', data=cars)
plt.title('Price Distribution by Car Make')
plt.xticks(rotation=45) 
plt.show()
```

```python
make_count = cars.Make.value_counts().reset_index().rename(columns={'index':'Make','Make':'Count'})

# Count plot make
sns.barplot(x='Make', y='Count', data=make_count)
plt.title('Number of Cars per Make')
plt.show()
```

```python
make_price = cars.groupby('Make')['Price'].mean().reset_index().sort_values('Price', ascending=False)

# Plot make by price
sns.barplot(x='Make', y= 'Price', data = make_price)
plt.title('Average Price of Cars by Make')
plt.xticks(rotation=45)
plt.show()
```

```python
count_locations = cars['Location'].value_counts().reset_index().rename(columns={'index':'Location','Location':'Count'})
# Count plot cars in each location
sns.barplot(x='Location', y='Count', data=count_locations)
plt.title('Car Count per Location')
plt.xticks(rotation=90) 
plt.show()
```

```python
loc_by_price = cars.groupby('Location')['Price'].mean().reset_index().sort_values('Price', ascending=False)

# Bar plot to compare average car prices in different locations
sns.barplot(x='Location', y='Price', data=loc_by_price)
plt.title('Average Car Prices By Locations')
plt.xticks(rotation=90) 
plt.show()
```

```python
# Count plot to visualize the distribution of car types
kind_count = cars.Kind.value_counts().reset_index()

# Plot count of kind
sns.barplot(x='index', y= 'Kind', data=kind_count)
plt.xlabel('Kind')
plt.ylabel('Count')
plt.title('Distribution of Car Types')
plt.xticks(rotation=45)
plt.show()
```

```python
kind_price = cars.groupby('Kind')['Price'].mean().reset_index().sort_values('Price', ascending=False)
# Plot kind by price
sns.barplot(x='Kind', y= 'Price', data=kind_price)
plt.title('Average Car Types by Price')
plt.xticks(rotation=45)
plt.show()
```

```python
count_model = cars.Model.value_counts().reset_index()

# Plot count of model
sns.barplot(x='index', y= 'Model', data=count_model)
plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Number of Cars per Model')
plt.xticks(rotation=90)
plt.show()
```

```python
model_price = cars.groupby('Model')['Price'].mean().reset_index().sort_values('Price', ascending=False)

plt.figure(figsize=(12,6))
# Plot bar plot of model by price
sns.barplot(x='Model', y= 'Price', data=model_price)
plt.title('Average Price of Cars by Model')
plt.xticks(rotation=90)
plt.show()
```

```python
doors_price = cars.groupby('Doors')['Price'].mean().reset_index()

# Bar plot of doors by price
sns.barplot(x='Doors', y= 'Price', data=doors_price)
plt.title('Average Price of Cars by Doors')
plt.show()
```

```python
make_age = cars.groupby('Make')['Age'].mean().reset_index().sort_values('Age', ascending=False)

# Bar plot of make by age
sns.barplot(x='Make', y= 'Age', data=make_age)
plt.ylabel('Age of Car (years)')
plt.title('Average Age of Cars by Make')
plt.show()
```

```python
kind_age = cars.groupby('Kind')['Age'].mean().reset_index().sort_values('Age', ascending=False)

# Bar plot of kind by age
sns.barplot(x='Kind', y= 'Age', data=kind_age)
plt.ylabel('Age of Car (years)')
plt.title('Average Age of Cars by Kind')
plt.show()
```

```python
model_age = cars.groupby('Model')['Age'].mean().reset_index().sort_values('Age', ascending=False)

# Bar plot of kind by age
sns.barplot(x='Model', y= 'Age', data=model_age)
plt.ylabel('Age of Car (years)')
plt.title('Average Age of Cars by Model')
plt.xticks(rotation=90)
plt.show()
```

```python
price_age = cars.groupby('Age')['Price'].mean().reset_index().sort_values('Price', ascending=False)

# Lineplot of Age and Price
sns.lineplot(x='Age', y='Price', data=price_age)
plt.title('Car Price vs. Age')
plt.xlabel('Age of Car (years)')
plt.ylabel('Price')
plt.show()
```

```python
price_year= cars.groupby('Year')['Price'].mean().reset_index().sort_values('Price', ascending=False)

# Lineplot of Age and Price
sns.lineplot(x='Year', y='Price', data=price_year)
plt.title('Car Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
# Heatmap 
corr_matrix = cars[['Year', 'Kms', 'Price', 'Doors','Age']].corr(numeric_only=False)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation plot')
plt.show()
```
